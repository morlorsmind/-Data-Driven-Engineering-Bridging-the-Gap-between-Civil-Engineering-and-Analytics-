```python
Lxp.isna().sum()
```

```python
LxpC = Lxp.dropna(subset=["Education Expenditure %"])
```

```python
LxpC.isna().sum()
```

```python
LxpC.duplicated().sum()
```

```python
fig = px.scatter_geo(
    LxpC, 
    locations="Country Code", 
    color="Life Expectancy World Bank",
    size="Education Expenditure %",
    hover_name="Country Name",
    projection="natural earth",
    title="Impact of Expenditure on Health and Education on Life Expectancy"
)

# Customize the appearance
fig.update_geos(showcoastlines=True, coastlinecolor="Black", showland=True, landcolor="white")
fig.update_layout(geo=dict(
    showcoastlines=True,
))

# Show the plot
fig.show()
```

```python

LxpC = Lxp.dropna(subset=["Health Expenditure %"])

fig = px.scatter_geo(
    LxpC, 
    locations="Country Code", 
    color="Life Expectancy World Bank",
    size="Health Expenditure %",
    hover_name="Country Name",
    projection="natural earth",
    title="Impact of Expenditure on Health and Education on Life Expectancy"
)


fig.update_geos(showcoastlines=True, coastlinecolor="Black", showland=True, landcolor="white")
fig.update_layout(geo=dict(
    showcoastlines=True,
))
```

```python


correlation_health = LxpC['Health Expenditure %'].corr(LxpC['Life Expectancy World Bank'])
correlation_education = LxpC['Education Expenditure %'].corr(LxpC['Life Expectancy World Bank'])

print(f"Correlation between Health Expenditure and Life Expectancy: {correlation_health:.2f}")
print(f"Correlation between Education Expenditure and Life Expectancy: {correlation_education:.2f}")

```
